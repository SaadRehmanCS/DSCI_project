Data retrieval and wrangling: we want to be able to use this data and apply a knn model to determine the likeliness of a death event to the patient based on their body levels of (variables to be decided). Patients are individuals in an age range of 40-95 years

```r
library(tidyverse)
library(tidymodels)
```

```r
heart_data_raw <- read_csv("https://raw.githubusercontent.com/SaadRehmanCS/DSCI_project/master/heart_failure_clinical_records_dataset.csv") %>%
                mutate(DEATH_EVENT = as_factor(DEATH_EVENT)) %>%
                rename(death_event = DEATH_EVENT) 

heart_data_condition <- heart_data_raw %>%
                pivot_longer(cols=c(anaemia, diabetes, high_blood_pressure), names_to="medical_condition", values_to="patient_has_it") %>%
                filter(patient_has_it == 1, medical_condition != "anaemia") %>%
                select(-patient_has_it, -medical_condition, -time, -sex, -smoking) 

heart_data <- heart_data_condition %>%
            filter(creatinine_phosphokinase < 1500)
                
nrow(heart_data)
head(heart_data)
```

We have now cleaned the data so it only includes patients who are already suffering from diabetes and high blood pressure. 
Next, we want to determine which variables have the greatest influence on death_event, and only use those variables in our actual knn model. We will do this by plotting several scatter plots of age vs *, where * includes creatinine_phospokinase:serum_sodium 

```r
creatine_plot <- heart_data %>%
                ggplot(aes(x = creatinine_phosphokinase, color = death_event)) +
                    geom_freqpoly(binwidth=20) + 
                    labs(x = "creatinine phosphokinase levels", y = "frequency of observations", color = "death event") +
                    theme(text = element_text(size=18))
creatine_plot
```

```r
ejection_plot <- heart_data %>%
                ggplot(aes(x = ejection_fraction, color = death_event)) +
                    geom_freqpoly(binwidth=1) + 
                    labs(x = "ejection fraction percentage", y = "frequency of observations", color = "death event") +
                    theme(text = element_text(size=18))
ejection_plot
```

```r
platelets_plot <- heart_data %>%
                mutate(platelets = platelets / 1000) %>%
                ggplot(aes(x = platelets, color = death_event)) +
                    geom_freqpoly(binwidth=15) + 
                    labs(x = "amount of platelets", y = "frequency of observations", color = "death event") +
                    theme(text = element_text(size=18))
platelets_plot
```

```r
serum_creatine_plot <- heart_data %>%
                ggplot(aes(x = serum_creatinine, color = death_event)) +
                    geom_freqpoly(binwidth=0.1) + 
                    labs(x = "serum creatinine levels", y = "frequency of observations", color = "death event") +
                    theme(text = element_text(size=18))
serum_creatine_plot
```

```r
sodium_plot <- heart_data %>%
                ggplot(aes(x = serum_sodium, color = death_event)) +
                    geom_freqpoly(binwidth=0.5) + 
                    labs(x = "serum sodium levels", y = "frequency of observations", color = "death event") +
                    theme(text = element_text(size=18))
sodium_plot
```

ejection_fraction, platelets, and both the serums have a visible relationship with death_event
